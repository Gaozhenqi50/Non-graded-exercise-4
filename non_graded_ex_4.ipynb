```python
import random
import time

# Define the cards with their attributes
cards = [
    {"name": "Diablo", "health": 100, "attack": 90, "defense": 60},
    {"name": "Medusa", "health": 100, "attack": 70, "defense": 70},
    {"name": "Jester", "health": 120, "attack": 60, "defense": 90},
    {"name": "Troll", "health": 150, "attack": 40, "defense": 94},
    {"name": "Specter", "health": 100, "attack": 70, "defense": 70},
    {"name": "Mist", "health": 100, "attack": 75, "defense": 65},
    {"name": "Savage", "health": 100, "attack": 90, "defense": 50},
    {"name": "Marauder", "health": 100, "attack": 85, "defense": 50},
    {"name": "Wimp", "health": 110, "attack": 40, "defense": 85},
    {"name": "Sorcerer", "health": 100, "attack": 70, "defense": 55}
]

# Function to assign a random deck of 5 cards to both player and computer
def assign_decks():
    deck = random.sample(cards, 5)
    return deck

# Function to display the cards in a player's deck
def display_deck(deck, owner="Player"):
    print(f"\n{owner}'s Deck:")
    for i, card in enumerate(deck):
        print(f"{i+1}. {card['name']} - Health: {card['health']}, Attack: {card['attack']}, Defense: {card['defense']}")
    print()

# Function to simulate an attack
def attack(player_card, opponent_card):
    # Player attacks opponent's card
    damage_to_defense = min(player_card['attack'], opponent_card['defense'])
    opponent_card['defense'] -= damage_to_defense
    print(f"{player_card['name']} attacks {opponent_card['name']}! {opponent_card['name']}'s defense decreases by {damage_to_defense} points.")

    if opponent_card['defense'] <= 0:
        # Remaining attack damages health
        damage_to_health = max(0, player_card['attack'] - damage_to_defense)
        opponent_card['health'] -= damage_to_health
        print(f"{opponent_card['name']}'s health decreases by {damage_to_health} points.")

    # Remove card if health drops to 0 or below
    if opponent_card['health'] <= 0:
        print(f"{opponent_card['name']} is defeated!")
        return True  # Card is defeated
    return False  # Card is still alive

# Game function
def play_game():
    player_deck = assign_decks()
    computer_deck = assign_decks()

    # Game loop
    while player_deck and computer_deck:
        display_deck(player_deck, "Player")
        
        # Computer selects a random card
        opponent_card = random.choice(computer_deck)
        print(f"Opponent's card is {opponent_card['name']} - Health: {opponent_card['health']}, Attack: {opponent_card['attack']}, Defense: {opponent_card['defense']}")
        
        # Player chooses a card
        choice = int(input(f"Choose a card from your deck (1-{len(player_deck)}): ")) - 1
        player_card = player_deck[choice]
        
        # Player's attack phase
        print("\nPlayer's Turn:")
        if attack(player_card, opponent_card):
            computer_deck.remove(opponent_card)
        time.sleep(3)  # Pause for effect
        
        # Check if computer has any cards left
        if not computer_deck:
            print("All opponent's cards have been defeated! You win!")
            break
        
        # Opponent's turn
        print("\nOpponent's Turn:")
        time.sleep(3)  # Pause before opponent attacks
        
        opponent_card = random.choice(computer_deck)
        display_deck(player_deck, "Player")
        choice = int(input(f"Choose a card from your deck (1-{len(player_deck)}): ")) - 1
        player_card = player_deck[choice]

        if attack(opponent_card, player_card):
            player_deck.remove(player_card)
        time.sleep(3)  # Pause for effect

        # Check if player has any cards left
        if not player_deck:
            print("All your cards have been defeated! Opponent wins!")
            break

# Start the game
play_game()
```
